```python
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go
import matplotlib.pyplot as plt

df = pd.read_csv('vBitIncidenciass.csv')

# Paso 2: Convertir la columna FechaInicio al formato de fecha
df['FechaInicio'] = pd.to_datetime(df['FechaInicio'], format='%d/%m/%Y')

# Paso 3: Extraer el mes de la columna FechaInicio
df['Mes'] = df['FechaInicio'].dt.strftime('%B')

# Paso 4: Agrupar por mes y tipo de incidencia, y contar las ocurrencias
data_incidentes = df.groupby(['Mes', 'Nombre']).size().reset_index(name='Count')

# Crear un gráfico de barras para cada mes
for mes in data_incidentes['Mes'].unique():
    mes_data = data_incidentes[data_incidentes['Mes'] == mes]
    plt.bar(mes_data['Nombre'], mes_data['Count'])
    plt.xlabel('Tipo de Incidencia')
    plt.ylabel('Cantidad de Incidentes')
    plt.title(f'Incidentes en {mes}')
    plt.show()
```
